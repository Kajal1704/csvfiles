```python
# Statistical analysis of classiifers for CGR- RY representation input files
import numpy as np
import scipy
from scipy.stats import binom

#Contingency table values- b and c for CGR-RY reprstn method: Tree and DNN
b=0 # b is the number of observations correctly classiifed by the first classifier but not the second
c=3 # c is the number of observations correctly classiifed by the second classifier but not the first
print("Statistical analysis for CGR-RY method: Tree and DNN")
print("b=",b)
print("c=",c)
n=b+c
print("n=",n)
x=min(b,c)
print("x=",x)
dist=binom(n, 0.5)
p=2. * dist.cdf(x)
print("Exact_p=",p)
midp=p-dist.pmf(x)
print("midp=",midp)
alpha=0.05
if p>alpha:
    print('Same proportions of error (fail to reject H0) i.e. there is no significant difference between misclassification of Tree and DNN')
else:
    print('Different proportions of errors (reject H0) i.e. there is significant difference between misclassification of Tree and DNN')
    
#Contingency table values- b and c for CGR-RY reprstn method: 1.SVM and DNN 2.RF and DNN 3.NN and DNN
b1=0 # b is the number of observations correctly classiifed by the first classifier but not the second
c1=0 # c is the number of observations correctly classiifed by the second classifier but not the first
print("\n Statistical analysis for CGR-RY method: 1.SVM and DNN 2.RF and DNN 3.NN and DNN")
print("b1=",b1)
print("c1=",c1)
n1=b1+c1
print("n1=",n1)
x1=min(b1,c1)
print("x1=",x1)
dist1=binom(n1, 0.5)
p1=2. * dist1.cdf(x1)
print("Exact_p1=",p1)
midp1=p1-dist1.pmf(x1)
print("midp1=",midp1)
alpha=0.05
if p1>alpha:
    print('Same proportions of error (fail to reject H0) i.e. there is no significant difference between misclassification of 1.SVM and DNN 2.RF and DNN and 3.NN and DNN')
else:
    print('Different proportions of errors (reject H0) i.e. there is significant difference between misclassification of 1.SVM and DNN 2.RF and DNN  3.NN and DNN')
```
